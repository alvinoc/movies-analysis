```python
import matplotlib.pyplot as plt 
import pandas as pd 
```

```python
movies_path = "C://Users//claro.alvino//Desktop//analise movies//core_movie.csv"
movies = pd.read_csv(movies_path,sep=',')
movies.head()
```

```python
movies_anos = movies['year']
movies_anos = [int(i) for i in movies_anos]
```

```python
is_american = movies['country'] == 'USA'
movies_american = movies[is_american]
print(movies_american.shape)
```

```python
budget = movies_american['budget'].tolist()
```

```python
print(budget)
```

```python
[i.strip('$') if type(i)== str else str(i) for i in budget]
```

```python
new_budget = [i.strip('$ PYG CAD RUR GBP') if type(i)== str else str(i) for i in budget]
```

```python
print(new_budget)
```

```python
new_budget_int = [i.strip(' nan ') if type(i)== str else str(i) for i in new_budget]
```

```python
new_budget_int
```

```python
no_whitespace = [i for i in new_budget_int if i.strip()]
```

```python
print(no_whitespace)
```

```python
budget_int = [i.strip('EUR') if type(i)== str else str(i) for i in no_whitespace]
```

```python
new_budget_int = [int(i) for i in budget_int]
```

```python
print(new_budget_int)
```

```python
plt.boxplot(new_budget_int)
plt.title("grafico ano x orcamento")
plt.show()
```
